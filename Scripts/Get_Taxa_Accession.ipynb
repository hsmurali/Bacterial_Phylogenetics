```python
import pandas as pd

core_genomes = eval(open('/Users/harihara/Mount/CMSC829A/Data/Core_Genome_List.txt').read())
df_accession = pd.read_csv('/Users/harihara/Mount/CMSC829A/Data/assembly_summary_bacteria.txt', 
                           skiprows=1, sep = '\t', index_col =['# assembly_accession'], low_memory = False)
df_taxa = pd.read_csv('ncbi_lineages_2021-12-03.csv.gz', low_memory = False, index_col = 'tax_id')
```

```python
filter_rows = []
for c in core_genomes:
    t = c.split('.')
    filter_rows.append(t[0]+'.'+t[1][0])
df_temp = pd.DataFrame(data = {'Tree_Leaves':core_genomes, 'Acc':filter_rows})
df_temp = df_temp.set_index('Acc')
df_taxa = df_taxa[['phylum','class','order','family','genus', 'species']]
df_accession_filter = df_accession.loc[filter_rows, ['taxid']]
df_accession_filter = df_accession_filter.reset_index().set_index('taxid')
df_accession_filter = df_accession_filter.join(df_taxa)
df_accession_filter.index.name = 'TaxID'
df_accession_filter = df_accession_filter.reset_index()
df_accession_filter = df_accession_filter.set_index('# assembly_accession').join(df_temp)
df_accession_filter = df_accession_filter.set_index('Tree_Leaves')
```

```python
df_accession_filter.to_csv('/Users/harihara/Mount/CMSC829A/Data/Taxonomy_Core_Genomes.txt')
```
