```python
from os import listdir, mkdir
import pandas as pd
from os.path import isdir
```

```python
filedir = '/Users/harihara/UMD_Coursework/Fall_2021/CMSC829A/Project/Data/Processed_GBFF/'
outprot = '/Users/harihara/Mount/CMSC829A/Data/Untagged_Accessory_Genes_Proteins/'
outnuc  = '/Users/harihara/Mount/CMSC829A/Data/Untagged_Accessory_Genes_Nucleotides/'

if not isdir(outprot):
    mkdir(outprot)

if not isdir(outnuc):
    mkdir(outnuc)

core_genomes = eval(open('/Users/harihara/Mount/CMSC829A/Data/Core_Genome_List.txt','r').read())
```

```python
msa_files = listdir(filedir)
for f in core_genomes:
    f = f+'_genomic.txt.gz'
    if f[0] != '.':
        print(f)
        df = pd.read_csv(filedir + f, sep = '\t',usecols = ['Start','End','locus_tag','gene',
                                                            'translation','DNA'])
        df.loc[df['gene'].isna(), 'gene'] = df.loc[df['gene'].isna(), 'locus_tag']
        genes = df.loc[df['gene'] == df['locus_tag'], 'gene'].tolist()
        proteins = df.loc[df['gene'] == df['locus_tag'], 'translation'].tolist()
        nuc = df.loc[df['gene'] == df['locus_tag'], 'DNA'].tolist()

        prot_file = outprot+f.replace(".txt.gz",".faa")
        nuc_file = outnuc+f.replace(".txt.gz",".fna")
        prot_buf = open(prot_file, 'w')
        nuc_buf = open(nuc_file, 'w')

        for i in range(len(genes)):
            prot_buf.write('>'+genes[i]+'\n')
            prot_buf.write(proteins[i]+'\n')

            nuc_buf.write('>'+genes[i]+'\n')
            nuc_buf.write(nuc[i]+'\n')
        prot_buf.close()
        nuc_buf.close()
            
```
