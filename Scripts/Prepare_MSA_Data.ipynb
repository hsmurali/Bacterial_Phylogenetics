```python
import pandas as pd
from os import listdir, mkdir
from os.path import isdir
data_path = 'UMD_Coursework/Fall_2021/CMSC829A/Project/Data/'
core_genomes = eval(open(data_path+'Core_Genome_List.txt','r').read())
core_genes = eval(open(data_path+'Core_Genes_List.txt','r').read())

prot_out_dir = data_path+'/Core_Gene_Proteins/'
if not isdir(prot_out_dir):
    mkdir(prot_out_dir)
    
nuc_out_dir = data_path+'/Core_Gene_Nucleotides/'
if not isdir(nuc_out_dir):
    mkdir(nuc_out_dir)
```

```python
dict_file_buf = {}
#Processed_GBFF/GCF_000005845.2_ASM584v2_genomic.txt.gz
for genome in core_genomes:
    print(genome)
    df = pd.read_csv(data_path+'Processed_GBFF/'+genome+'_genomic.txt.gz', sep = '\t')
    del df['Unnamed: 0']
    df = df[df['gene'].isin(core_genes)]
    genes = df['gene'].tolist()
    proteins = df['translation'].tolist()
    nuc_seqs = df['DNA'].tolist()
    for i in range(0, len(genes)):
        gene = genes[i]
        prot = proteins[i]
        nuc = nuc_seqs[i]
        try:
            dict_file_buf[gene][0].write('>'+genome+'\n')
            dict_file_buf[gene][0].write(prot+'\n')
            dict_file_buf[gene][1].write('>'+genome+'\n')
            dict_file_buf[gene][1].write(nuc+'\n')
        except KeyError:
            dict_file_buf[gene] = []
            dict_file_buf[gene].append(open(prot_out_dir+gene+'.faa','w'))
            dict_file_buf[gene].append(open(nuc_out_dir+gene+'.fna','w'))
            dict_file_buf[gene][0].write('>'+genome+'\n')
            dict_file_buf[gene][0].write(prot+'\n')
            dict_file_buf[gene][1].write('>'+genome+'\n')
            dict_file_buf[gene][1].write(nuc+'\n')

for d in dict_file_buf:
    print(d)
    dict_file_buf[d][0].close()
    dict_file_buf[d][0].close()
    dict_file_buf[d][1].close()
    dict_file_buf[d][1].close()
            
```
